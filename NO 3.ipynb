```python
import pandas as pd

# Load the dataset
df = pd.read_csv("iris.csv")

# Filter the DataFrame for specific species
species_data = df[df['species'].isin(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])]

# Calculate basic statistical details for the filtered species
species_stats = species_data.groupby('species').describe()

print(species_stats)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data.csv")

# Group by the categorical variable and calculate summary statistics for the numeric variable
summary_stats = df.groupby("category")["income"].describe()

# Create a list of summary statistics for each category
summary_list = summary_stats.values.tolist()

print(summary_stats)
print(summary_list)
```
